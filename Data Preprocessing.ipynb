```python
import pandas as pd
```

```python
Player=pd.read_excel(r'C:\Users\USER\OneDrive\Documents\ds_c8_datafiles-main\players.xlsx')
Game=pd.read_excel(r'C:\Users\USER\OneDrive\Documents\ds_c8_datafiles-main\games.xlsx')
Game_event=pd.read_excel(r'C:\Users\USER\OneDrive\Documents\ds_c8_datafiles-main\game_events.xlsx')
Game_line=pd.read_excel(r'C:\Users\USER\OneDrive\Documents\ds_c8_datafiles-main\game_lineups.xlsx')
Appe=pd.read_excel(r'C:\Users\USER\OneDrive\Documents\ds_c8_datafiles-main\appearances.xlsx')
```

```python
FootBall_data=pd.merge(Appe,Game_event,on='game_id',how='outer')
FootBall_data=pd.merge(FootBall_data,Game_line,on='game_id',how='outer')
FootBall_data=pd.merge(FootBall_data,Game,on='game_id',how='outer')
FootBall_data=pd.merge(FootBall_data,Player,on='player_id',how='outer')
```

```python
FootBall_data
```

```python
FootBall_data.isnull().sum()
```

```python
for col in FootBall_data.columns:
    if FootBall_data[col].dtypes=='object':
        FootBall_data[col].fillna(FootBall_data[col].mode()[0],inplace=True)
    else:
        FootBall_data[col].fillna(FootBall_data[col].mean(),inplace=True)
```

```python
FootBall_data.isnull().sum()
```

```python
FootBall_data.drop(columns=['home_club_formation'],inplace=True)
FootBall_data.drop(columns=['away_club_formation'],inplace=True)
```
